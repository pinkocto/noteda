## Values are not uniquely identified 에러

> SOLAR 데이터 재구조화 과정에서 발생한 오류였음.

](attachment:53a81c71-ea80-453f-85d8-b08426401835.png)

- value가 유니크하지 않아서 발생하는 에러이다. 각 region별 임의로 유니크한 식별 로우를 만들어 주면 에러를 해결할 수 있음.

## ***Solution***

```python
df %>%
  group_by(region) %>%
  mutate(row = row_number()) %>%
  tidyr::pivot_wider(names_from = region, values_from = solar_radiation) %>%
  select(-row)
```
