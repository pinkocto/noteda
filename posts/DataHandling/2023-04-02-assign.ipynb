## 방법3: .assign으로 할당 ($\star$)
판다스로 새로운 열 할당하는 방법에는 여러 방법들이 있지만 그 중 assign으로 할당하는 것은 확장성이 있고 다양한 상황에 사용하기 좋음.

```python
import pandas as pd
```

```python
df = pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df.assign(c=[3,4,5])
```

```python
df.assign(c=[3,4,5],d=[4,5,6])
```

```python
df.assign(c=[3,4,5]).assign(d=[4,5,6])
```

### 장점1. 원래 데이터프레임 그대로 유지.

```python
df.assign(c=[3,4,5]).assign(d=[4,5,6])
```

```python
df # 원래 데이터 프레임은 그대로 유지된다.
```

### 장점2. 수정 용이

```python
df.assign(c=[3,4,5]).assign(d=[4,5,6]) # 1->2, 2->3 으로 가는 과정이 메모리 공간안에 모두 저장되어 있다.
```

```python
df ## step1
```

```python
df.assign(c=[3,4,5]) ## step2
```

```python
df.assign(c=[3,4,5]).assign(d=[4,5,6]) ## step3
```

```python
## d=[4,5,6] >> d = [4,-5,6]으로 변환.
df.assign(c=[3,4,5]).assign(d=[4,-5,6]) 
```

## 예시1

`color ~ num_voted_user`를 뽑고 + `aspect_ratio`도 추가적으로 뽑으시오.

### `-` 안되는 예

```python
df.loc[:,['color':'num_voted_users','aspect_ratio']] # 이건 안됨.
```

### `-` 해결 (iloc 이용)

***Tip!*** :  복잡한 조건은 iloc으로 쓰는게 편할때가 있다. 
 $\to$ 그런데 `df.columns` 변수들이 몇번인지 알아보기 힘듬 $\to$
 아래와 같이 하면 열의 이름을 인덱스와 함께 출력할 수 있음

```python
pd.Series(df.columns) # 매우 편함.
```

- 해당 열의 인덱스와 컬럼명이 같이 출력된다. 일일이 세지 않아도 된다.

```python
list(range(13))+[26]
```

```python
df.iloc[:,list(range(13))+[26]].head(2)
```
