```r
library(data.table)
library(tidyverse)
library(dplyr)
library(lubridate)
library(recipes)
library(corrplot)
library(GGally)
```

```r
url = 'https://raw.githubusercontent.com/pinkocto/mm/main/posts/2_Research/SOLAR/solar_radiation.csv'
df = read_csv(url)
```

```r
df = df |> mutate(date=ymd_hm(date))
```

```r
df%>%colnames
```

```r
df <- df %>%
          group_by(region) %>%
          mutate(row = row_number()) %>%
          tidyr::pivot_wider(names_from = region, values_from = solar_radiation) %>%
          select(-row)
```

```r
head(df)
```

## Correlation

```r
num_vars <- df %>% select(-date)
weight <- cor(num_vars)
weight
```

## Corrplot

```r
options(repr.plot.width=30, repr.plot.height=30, repr.plot.res=300)
```

```r
corrplot(round(weight,2), method = 'number')
```
