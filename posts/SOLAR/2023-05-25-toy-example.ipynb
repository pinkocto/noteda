# Toy Example

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = np.linspace(0,1,96)
noise = np.random.normal(0, 0.3, len(t))
x1 = np.sin(4*np.pi*t) + 2
x2 = x1 + noise
x3 = np.cos(4*np.pi*t) + 2
x4 = x3 + noise
plt.figure(figsize=(10,3))
plt.plot(t,x1, color='blue')
plt.plot(t,x2, color='blue')
plt.plot(t,x3, color='red')
plt.plot(t,x4, color='red')
plt.show()
```

```python
X = pd.DataFrame(np.ones((96,4)))
X.iloc[:,0]=x1
X.iloc[:,1]=x2
X.iloc[:,2]=x3
X.iloc[:,3]=x4
```

```python
X = X.rename(columns={0:'node1', 1:'node2', 2:'node3', 3:'node4'})
```

# SIMULATION

```python
plans_stgcn = {
    'max_iteration': 50,  # 30, 
    'method': ['Ver1'], 
    'lags': [4, 8, 24], #[4, 8, 12, 24], 
    'nof_filters':[4, 16, 32], # [16, 32, 64], 
    'epoch': [50] # [50, 100, 150]
}
```

```python
plans_stgcn2 = {
    'max_iteration': 50,  # 30, 
    'method': ['Ver2'], 
    'lags': [4, 8, 24], #[4, 8, 12, 24], 
    'nof_filters':[4, 16, 32], # [16, 32, 64], 
    'epoch': [50] # [50, 100, 150]
}
```

```python
plans_stgcn
```

```python
import eptstgcn.planner
plnr = eptstgcn.planner.PLNR_STGCN(plans_stgcn,loader1,dataset_name='ver1')
plnr2 = eptstgcn.planner.PLNR_STGCN(plans_stgcn2,loader2,dataset_name='ver2')
```

```python
plnr.simulate()
eptstgcn.save_data(plnr, './toy_example/model/ver1_50.pickle')
simul_model1 = eptstgcn.load_data('./toy_example/model/ver1_50.pickle')

plnr2.simulate()
eptstgcn.save_data(plnr2, './toy_example/model/ver2_50.pickle')
simul_model2 = eptstgcn.load_data('./toy_example/model/ver2_50.pickle')
```

```python
# plnr2.simulate()
# eptstgcn.save_data(plnr2, './toy_example/model/ver2.pickle')
simul_model2 = eptstgcn.load_data('./toy_example/model/ver2.pickle')
```

```python
v1_50 = simul_model1.simulation_results
v2_50 = simul_model2.simulation_results
df = pd.concat([v1_50, v2_50])
```

## SIMUL RSLT

```python
df.groupby(["dataset","lags","nof_filters"])\
.agg({"mse(train)":["mean"], "mse(test)":["mean"]})
```
