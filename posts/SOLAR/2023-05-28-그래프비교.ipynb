```python
def CompareGraph(region_lst, n, k):
    color_lst = ['red','blue','purple','darkgreen']
    color_=random.choice(color_lst)
    l = len(region_lst)
    plt.figure(figsize=(10,6))
    for i in range(l):
        plt.subplot(l,1,i+1)
        # plt.plot(range(n), solar[region_lst[i]][:n] , '-', color=color_, alpha=.7)
        plt.plot(range(abs(k),abs(n)), solar[region_lst[i]].iloc[n:k] , '-', color=color_, alpha=.7)
        plt.title('Graph{}: {}'.format(i+1, region_lst[i]))
        plt.ylabel('solar radiation')
        
    plt.tight_layout()
    plt.show()
```
