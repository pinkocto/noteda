```r
library(tidyverse)
library(lubridate)
library(EPT)
```

```r
url = 'https://raw.githubusercontent.com/miruetoto/yechan3/main/posts/3_Researches/SOLAR/solar_radiation.csv'
df = read_csv(url)
```

```r
df = df |> filter(region == '북춘천') |> mutate(date=ymd_hm(date))
df
```

`-` 지역을 북춘천으로 고정 

```r
df2 = df |> filter(region =='북춘천') 
df2 = df2[order(df2$date),]
```

```r
y = df2$solar_radiation
y
```

```r
plot(y[1:500])
lines(y[1:500],lty=2)
```

`-` EPT 수행 

```r
ept = function(y){
    EpM = eptransf(signal=y,tau=24,process=c("envelope","average"))$EpM
    EpM*2
}
```

```r
yU = ept(y)
```

```r
plot(y[1:500])
lines(yU[1:500],col=2,lty=2)
```
