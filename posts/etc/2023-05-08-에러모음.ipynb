##  align error

- <https://tex.stackexchange.com/questions/256920/package-amsmath-error-beginaligned-allowed-only-in-math-mode>

두줄 이상은 {aligned} 사용해야함. (align*, align이런 것들 안됨, equation, split 이런것들도 안됐음.)

패키지 amsmath 오류: \begin{aligned}는 수학 모드에서만 허용됩니다.

LaTeX는 수학 표현식에 대한 두 가지 쓰기 모드인 인라인 수학 모드와 디스플레이 수학 모드를 허용합니다.

인라인 수학 모드는 단락의 일부인 수식을 작성하는 데 사용됩니다.
디스플레이 수학 모드는 단락의 일부가 아닌 식을 작성하는 데 사용되므로 별도의 줄에 표시됩니다.
인라인 수학 모드 구분 기호

```
\(...\)
$...$
\begin{math}...\end{math}
```
수학 모드 구분 기호 표시
```
\[...\]
\begin{displaymath}...\end{displaymath}
\begin{equation}...\end{equation}
```
한마디로 \begin{aligned}이러한 수학 모드 구분 기호 중 하나에 작성해야 합니다.
